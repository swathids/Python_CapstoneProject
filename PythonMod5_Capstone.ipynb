```python
#Preprocessing: Correct the data in the "height" column by replacing it with random numbers between 150 and 180. Ensure data consistency and integrity before proceeding with 

import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("ABC.csv")
data.info()
```

```python
print(data.head())
data['Height']=np.random.randint(150,181,size=len(data))
print(data.head())
```

```python
#Here the dataset "ABC.csv" is loaded using pd.read_csv(), and an overview is obtained with data.info() and data.head(). To correct the "Height" column, missing or invalid values are replaced with random integers between 150 and 180 using np.random.randint(). Finally, print(data.head()) is used to verify the updates, ensuring the data is ready for analysis.
```

```python
# Q1.Determine the distribution of employees across each team and calculate the percentage split relative to the total number of employees.data

Team_dist=data['Team'].value_counts()
percent_split=(Team_dist/len(data))*100
dist_summary=pd.DataFrame({
    'Team':Team_dist.index,
    'Emp_count':Team_dist.values,
    'Percentage split':percent_split
})

dist_summary.reset_index(drop=True,inplace=True)
print(dist_summary)


# calculates the employee count and percentage distribution across teams, providing insights into team sizes relative to the total workforce.




```

```python
#Q2.Segregate employees based on their positions within the company. 

positions=data['Position'].unique()
segregated_data={position:data[data['Position']==position]['Name'].tolist() for position in positions}
for i in segregated_data.keys():
    print(/n" Employees in position",i)
    print(", ".join(segregated_data[i]))

#segregates employees based on their positions and lists the names of employees in each position, providing an overview of how employees are distributed across different roles.
```

```python
#Q3.Identify the predominant age group among employees

sort_data=data.sort_values(by='Age')
sort_data
```

```python
#grouping the data
bins=[19,24,29,34,39,44]
labels=['19-24','25-29','30-34','35-39','40-44']
data['Agegroup']=pd.cut(data['Age'],bins=bins,labels=labels,right=False)
Agegroup_dist=data['Agegroup'].value_counts()
predom_grp=Agegroup_dist.idxmax()
predom_count=Agegroup_dist.max()
print(f"Predominant Age Group:{predom_grp}({predom_count} Epmloyees)")


#Groups employees into age ranges, chosen as 19-24, 25-29, 30-34, 35-39, and 40-44 for simplicity, counts employees in each range, and identifies the most common age group with its count.
#Identifies the predominant age group with the highest count, displaying the result.
```

```python
#Q4. Discover which team and position have the highest salary expenditure.

#calculates the total salary expenditure for each team and position, identifies the team and position with the highest expenditure, and displays the results.
#It helps pinpoint where the organization's salary costs are concentrated.


Sal_exp_team=data.groupby('Team')['Salary'].sum().reset_index()
High_team=Sal_exp_team[Sal_exp_team['Salary']==Sal_exp_team['Salary'].max()]
High_exp_team=Sal_exp_team['Salary'].max()
print(f"Team with highest salary expenditure: {High_team.iloc[0]['Team']}, Salary expenditure: {High_exp_team}")


Sal_exp_pos=data.groupby('Position')['Salary'].sum().reset_index()
High_pos=Sal_exp_pos[Sal_exp_pos['Salary']==Sal_exp_pos['Salary'].max()]
High_exp_pos=Sal_exp_pos['Salary'].max()
print(f"Position with highest salary expenditure: {High_pos.iloc[0]['Position']}, Salary expenditure: {High_exp_pos}")

```

```python
# Q5.Investigate if there's any correlation between age and salary, and represent it visually.

# computes the correlation between age and salary. 
# visually represents the relationship using a scatterplot with a regression line, providing insights into how age and salary might be related.


import matplotlib.pyplot as plt
import seaborn as sns
correlation=data['Age'].corr(data['Salary'])
print(f"Correlation between Age and Salary: {correlation:.2f}")
plt.figure(figsize=(8,6))
sns.scatterplot(x='Age',y='Salary',data=data,color='green',alpha=0.6)
sns.regplot(x='Age',y='Salary',data=data,scatter=False,color='red')
plt.title('Age vs Salary Correlation')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
# Plot 1: Employee Distribution Across Teams (Bar Plot)
plt.figure(figsize=(12, 6))
sns.barplot(x='Team', y='Emp_count', data=dist_summary, palette='viridis')
plt.title("Employee Distribution Across Teams", fontsize=16)
plt.xlabel("Team", fontsize=12)
plt.ylabel("Number of Employees", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot 2: Percentage Split Across Teams (Bar Plot)
plt.figure(figsize=(12, 6))
sns.barplot(x='Team', y='Percentage split', data=dist_summary, palette='Blues')
plt.title("Percentage Split of Employees Across Teams", fontsize=16)
plt.xlabel("Team", fontsize=12)
plt.ylabel("Percentage Split (%)", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Employee Distribution Across Teams: This bar plot shows the number of employees in each team.
#Percentage Split Across Teams: This bar plot illustrates the percentage of employees in each team relative to the total workforce.
```

```python
# Count the number of employees in each position
position_count = data['Position'].value_counts()

# Plotting the distribution of employees by position (Bar Plot)
plt.figure(figsize=(12, 6))
sns.barplot(x=position_count.index, y=position_count.values)
plt.title("Number of Employees in Each Position", fontsize=16)
plt.xlabel("Position", fontsize=12)
plt.ylabel("Number of Employees", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.tight_layout()
plt.show()
```

```python
#Q4. Define age bins and labels
bins = [19, 24, 29, 34, 39, 44]  # Adjust these values as per your data
labels = ['19-24', '25-29', '30-34', '35-39', '40-44']

# Categorize employees into age groups
data['Agegroup'] = pd.cut(data['Age'], bins=bins, labels=labels, right=False)

# Count the number of employees in each age group
age_group_count = data['Agegroup'].value_counts()

# Plotting the distribution of employees across age groups
plt.figure(figsize=(12, 6))
sns.barplot(x=age_group_count.index, y=age_group_count.values)
plt.title("Employee Distribution Across Age Groups", fontsize=16)
plt.xlabel("Age Group", fontsize=12)
plt.ylabel("Number of Employees", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


```

```python
#This code categorizes employees into different age groups based on predefined age bins and then visualizes the distribution of employees across these groups.
#The bar plot shows how many employees fall into each age group, providing insights into the predominant age demographics within the company.
```

```python
# Q5.Grouping the data by Team and Position, and summing up the Salary for each
Sal_exp_team = data.groupby('Team')['Salary'].sum().reset_index()
Sal_exp_pos = data.groupby('Position')['Salary'].sum().reset_index()

# Plot 1: Salary Expenditure by Team (Bar Plot)
plt.figure(figsize=(12, 6))
sns.barplot(x='Team', y='Salary', data=Sal_exp_team)
plt.title("Salary Expenditure by Team", fontsize=16)
plt.xlabel("Team", fontsize=12)
plt.ylabel("Total Salary Expenditure", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot 2: Salary Expenditure by Position (Bar Plot)
plt.figure(figsize=(12, 6))
sns.barplot(x='Position', y='Salary', data=Sal_exp_pos)
plt.title("Salary Expenditure by Position", fontsize=16)
plt.xlabel("Position", fontsize=12)
plt.ylabel("Total Salary Expenditure", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Salary Expenditure by Team: The first plot shows the total salary expenditure across different teams, highlighting which team has the highest financial commitment.

#Salary Expenditure by Position: The second plot displays salary expenditure by position, helping identify which roles incure the highest salary costs in the company.

```
